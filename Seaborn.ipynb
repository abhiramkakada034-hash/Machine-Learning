**What is Seaborn? (Beginner-Friendly Explanation)**

Seaborn is a Python data visualization library built on top of Matplotlib, designed to make beautiful, statistical plots with very little code.

Think of it like this:

Matplotlib = Manual car (more control, more effort)

Seaborn = Automatic car (cleaner, easier, more stylish)

Seaborn is especially powerful for data analysis because it can directly visualize:

**distributions**

**correlations**

**categories**

**relationships**

**trends**

# **Seabornâ€™s main plot categories**

# **Seaborn Plotting Functions**

The Seaborn library provides us with relplot() function and this function provides access to several different axes-level functions that show the relationship between two variables with semantic mappings of subsets. The kind parameter selects the underlying axes-level function to use:

scatterplot() (with kind="scatter")
lineplot() (with kind="line")

# **What is relplot()**

relplot() stands for RELational PLOT.

It is used to visualize relationships between variables, making it perfect for:

scatter plots

line plots

grouping

faceting/multi-plot layouts

Think of relplot() as a higher-level wrapper that can create multiple related plots at once.

```python
import seaborn as sns
tips = sns.load_dataset("tips")
tips.head()
```

```python
sns.relplot(data=tips, x="total_bill", y="tip")
sns.set_context("notebook")
sns.set()
```

```python
sns.relplot(data=tips, x="total_bill", y="tip", hue='day')
sns.set_context("notebook")
sns.set()
```

```python
sns.relplot(data=tips, x="total_bill", y="tip", hue='day', col='sex')
sns.set_context("notebook")
sns.set()
```

```python
sns.relplot(data=tips, x="total_bill", y="tip", hue='sex', col='day',col_wrap=2)
sns.set_context("notebook")
sns.set()
```

# **Histogram**

Histograms represent the data distribution by forming bins along with the range of the data and then drawing bars to show the number of observations that fall in each bin.In Seaborn we use displot() or histplot() function to plot histograms.

**When to use:**

âœ”ï¸ The shape of data distribution

âœ”ï¸ Check skewness

âœ”ï¸ Check if data is normal (bell curve)

âœ”ï¸ Compare distributions between groups

```python
import seaborn as sns
from matplotlib import pyplot as plt
df = sns.load_dataset('iris')
sns.displot(df['petal_length'],kde = True)
```

```python
tips = sns.load_dataset("tips")
sns.histplot(
    data=tips,
    x="total_bill",
    bins=25,
    kde=True,
    hue="sex",
    multiple="dodge",
    stat="density",
    element="bars",#step, poly
    palette="coolwarm"
)
plt.show()
#stat= Controls what the height of the bars represent
""" "count" â†’ raw count

"frequency"

"density" â†’ area under bars = 1

"probability" â†’ probability distribution

"percent" â†’ bars add to 100% """
```

# **Bar Plot**

# **What is sns.barplot()?**
barplot() is used to visualize the relationship between a categorical variable and a numerical variable by showing the mean (default) of the numerical variable for each category.

ðŸ“Œ Seaborn automatically calculates:

-> mean for each category

-> confidence interval (CI) around the mean

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_context('paper')

# load dataset
titanic = sns.load_dataset('titanic')
# create plot
sns.barplot(x = 'embark_town', y = 'age', data = titanic,
            palette = 'PuRd',ci=None ,hue='embark_town'
            )
plt.legend()
plt.show()
print(titanic.columns)
```

**ci= (Confidence Interval)**

Controls the error bar (vertical line).

Options:

ci=95 (default)

ci=None â†’ remove error bars

ci=68, etc.

```python
import matplotlib.pyplot as plt
import seaborn as sns
# load dataset
titanic = sns.load_dataset('titanic')
# create plot
sns.barplot(x = 'sex', y = 'survived', hue = 'class', data = titanic,
            palette = 'PuRd',
            order = ['male', 'female'],
            capsize = 0.05,
            saturation = 5,
            errcolor = 'blue',
            ci = 'sd'
            )
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
# load dataset
titanic = sns.load_dataset('titanic')
# create plot
sns.barplot(x = 'sex', y = 'survived', hue = 'class', data = titanic,
            palette = 'PuRd',
            order = ['male', 'female'],
            capsize = 0.05,
            saturation = 8,
            errcolor = 'gray',
            ci = 'sd'
            )
plt.legend()
plt.show()
```

**Use vertical barplots when:**

Categories are short

Time-based or naturally ordered data

Comparison of heights is intuitive

**Use horizontal barplots when:**

Labels are long

Many categories

Ranking and readability matter

# **Count Plot**

The count plot can be thought of as a histogram across a categorical variable.

countplot() is used to count the number of occurrences of each category in a categorical variable and visualize those counts as bars.

You donâ€™t provide the â€œvaluesâ€ â€”
Seaborn automatically counts them for you.

Think of it as:

âœ” â€œValue counts + barplot combined into one function.â€

**When do we use countplot()?**

âœ” You want to see how many times each category appears

âœ” You want to understand class balance

âœ” You want to analyze frequency distribution

```python
import matplotlib.pyplot as plt
import seaborn as sns
titanic = sns.load_dataset('titanic')
sns.countplot(x='class', data=titanic, hue='who', palette='magma')
sns.despine()
plt.title('Survivors')
plt.show()
```

# **Point Plot**

pointplot() shows the trend of category-wise averages, making comparisons and patterns very easy to see.


# **When do we use pointplot()?**

Use pointplot when you want to answer questions like:

âœ” How does the average value change across categories?

Example: Average salary across job roles

âœ” How do two categorical variables interact?

Example: Average exam score for each gender (category) and class level (hue)

âœ” You want a cleaner alternative to bar plots

Barplots can hide small differences; pointplots make trends easier to see.

âœ” You want to compare means with confidence intervals

Useful for statistical visualization.

```python
# importing required packages
import seaborn as sns
import matplotlib.pyplot as plt

# loading dataset
data = sns.load_dataset("tips")
sns.pointplot(x="day", y="tip", data=data)
plt.show()
```

```python
sns.pointplot(x="time", y="total_bill", hue="smoker",
                   data=data, palette="Accent")
```

# **Joint Plot**

jointplot() is a special type of plot that shows:

1ï¸âƒ£ The relationship between two variables (the â€œjointâ€ plot)

â†’ usually a scatter, hexbin, or KDE plot

2ï¸âƒ£ The distribution of each variable separately (the â€œmarginalâ€ plots)

â†’ histograms or KDE curves on the top and right side

So it is like combining scatterplot + histogram in one figure.

Because it visualizes:

Joint distribution â†’ how X and Y relate

Marginal distributions â†’ how X behaves alone, how Y behaves alone

No other single function does this combination so neatly.

# **When should you use jointplot()?**

Use it when you want to:

âœ” Understand relationship between two continuous variables

Examples: Height vs Weight,Age vs Salary,Study time vs Marks

âœ” See patterns: correlation, clusters, outliers

Jointplot makes these easy to spot visually.

âœ” Show both relationship AND individual distributions

Which is extremely useful in exploratory data analysis (EDA).

âœ” Present results in a clean, compact visualization

Looks great in reports and dashboards.

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_style("dark")
tips=sns.load_dataset('tips')
sns.jointplot(x='total_bill', y='tip',data=tips)
```

```python
# Add regression line to scatter plot and kernel density estimate to histogram
sns.jointplot(x='total_bill', y='tip', data=tips, kind='reg')
```

```python
# Display kernel density estimate instead of scatter plot and histogram
sns.jointplot(x='total_bill', y='tip', data=tips, kind='kde')
```

```python
# Display hexagons instead of points in scatter plot
sns.jointplot(x='total_bill', y='tip', data=tips, kind='hex')
```
